### **PANDAS PLOT**

Q1. Plot Artist wise mean of energy and liveness on a bar chart for the artist who have three or more song in the dataset.

```python
# Q1. Plot Artist wise mean of energy and liveness on a bar chart for the artist who have three or more song in the dataset.
```

```python
r1 = df['artist'].value_counts()
df_more3 = r1[r1 > 3]
morethan3 = df_more3[df_more3 >= 3].index
res1 = df[df['artist'].isin(morethan3)].groupby('artist')[['energy', 'liveness']].mean()
res1.plot.bar(stacked= True)
plt.grid()
plt.show()
```

```python
res1.plot.bar(subplots = True,color = {"energy":"maroon","liveness":"blue"})
plt.show()
```

Q2.Plot genre wise mean of danceability and valance on a line chart for the artist who have 3 or more song in the dataset.
    Display only top 7 mean of danceability based on genera

```python
r2 = df.groupby('top genre')[['danceability','valance']].mean()
r2 = r2.sort_values(by = 'danceability', ascending = False).head(6)
r2.plot.line(rot = 90, color = ['brown','green'])
plt.show()
```

Q3.Plot year wise mean of popularity on a pie chart for the year 2017 and above

```python
df_year_2017 = df[df['year'] >= 2017]
rf3 = df_year_2017.groupby('year')['popularity'].mean()
rf3.plot(kind='pie', autopct='%.2f%%', labeldistance=1.035)
plt.show()
```

 Q4.Plot a pie chart for the mean of popularity and energy for based on the year of the years 2017 and beyond

```python
r5 = df_year_2017.groupby('year')[['popularity','energy']].mean()
r5.plot.pie(
    autopct='%.2f%%',
    labeldistance=1.035,
    subplots=True,
    legend=False,
    figsize=(10, 5)
)
plt.show()
```

# Q5. Plot mean of Energy and danceability bases on different year



```python
r6 = df.groupby('year')[['energy','danceability']].mean()
r6.plot(x = 'energy', y = 'danceability',kind = 'scatter')
plt.show()
```
